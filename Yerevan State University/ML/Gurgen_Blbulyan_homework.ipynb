## a)
We have this PMF
$$f_{X,Y}(x,y)$$


|Y\X|-1|0|2|
|---|---|---|---|
|0|0.2|0.1|0|
|1|0.15|0.1|0.25|
|2|0.1|0|0.1|

From this PMF we can get $f_{X}(x)$ and $f_{Y}(y)$

$$f_{X}(x)$$

|X|-1|0|2|
|---|---|---|---|
|P(X=x)|0.45|0.2|0.35|

$$f_{Y}(y)$$

|Y|0|1|2|
|---|---|---|---|
|P(Y=y)|0.3|0.5|0.2|

To find $\mathbb{E}(X|Y)$ and $\mathbb{E}(Y|X)$ let's calculate conditional PMFs

$$f_{Y|x=-1}(y)$$

|Y\x=-1|0|1|2|
|---|---|---|---|
|P|$\frac{0.2}{0.45}$|$\frac{0.15}{0.45}$|$\frac{0.1}{0.45}$|

$$f_{Y|x=0}(y)$$

|Y\x=0|0|1|2|
|---|---|---|---|
|P|$\frac{0.1}{0.2}$|$\frac{0.1}{0.2}$|$\frac{0}{0.2}$|

$$f_{Y|x=2}(y)$$

|Y\x=2|0|1|2|
|---|---|---|---|
|P|$\frac{0}{0.35}$|$\frac{0.25}{0.35}$|$\frac{0.1}{0.35}$|

$$f_{X|y=0}(x)$$

|X\y=0|-1|0|2|
|---|---|---|---|
|P|$\frac{0.2}{0.3}$|$\frac{0.1}{0.3}$|$\frac{0}{0.3}$|

$$f_{X|y=1}(x)$$

|X\y=1|-1|0|2|
|---|---|---|---|
|P|$\frac{0.15}{0.5}$|$\frac{0.1}{0.5}$|$\frac{0.25}{0.5}$|

$$f_{X|y=2}(x)$$

|X\y=2|-1|0|2|
|---|---|---|---|
|P|$\frac{0.1}{0.2}$|$\frac{0}{0.2}$|$\frac{0.1}{0.2}$|







### 1.
$\mathbb{E}(Y|x=-1)=0\cdot\frac{0.2}{0.45}+1\cdot\frac{0.15}{0.45}+2\cdot\frac{0.1}{0.45}=\frac{0.35}{0.45}$

$\mathbb{E}(Y|x=0)=0\cdot\frac{0.1}{0.2}+1\cdot\frac{0.1}{0.2}+2\cdot\frac{0}{0.2}=\frac{0.1}{0.2}$

$\mathbb{E}(Y|x=2)=0\cdot\frac{0}{0.35}+1\cdot\frac{0.25}{0.35}+2\cdot\frac{0.1}{0.35}=\frac{0.45}{0.35}$

$\mathbb{E}(X|y=0)=-1\cdot\frac{0.2}{0.3}+0\cdot\frac{0.1}{0.3}+2\cdot\frac{0}{0.3}=-\frac{0.2}{0.3}$

$\mathbb{E}(X|y=1)=-1\cdot\frac{0.15}{0.5}+0\cdot\frac{0.1}{0.5}+2\cdot\frac{0.25}{0.5}=\frac{0.35}{0.5}$

$\mathbb{E}(X|y=2)=-1\cdot\frac{0.1}{0.2}+0\cdot\frac{0}{0.2}+2\cdot\frac{0.1}{0.2}=\frac{0.1}{0.2}$

### 2.
$\mathbb{E}(\mathbb{E}(Y|X))=\frac{0.35}{0.45}P(x=-1)+\frac{0.1}{0.2}P(x=0)+\frac{0.45}{0.35}P(x=2)=\frac{0.35}{0.45}\cdot0.45+\frac{0.1}{0.2}\cdot0.2+\frac{0.45}{0.35}\cdot0.35=0.9$

$\mathbb{E}(Y)=0\cdot0.3+1\cdot0.5+2\cdot0.2=0.9$

We get that $\mathbb{E}(\mathbb{E}(Y|X))=\mathbb{E}(Y)$

\\ 0, & else  \end{cases}

$$
f_{X}(x) = \begin{cases} \int_{0}^{x+1}1dy, & x\in[-1,0]
\\\int_{0}^{1-x}1dy, & x\in(0,1]

\\1-x, & x\in(0,1]
\\ 0, & \mbox{else } \end{cases}
$$

### 1.
$$f_{Y|X}(y|x)=\frac{f_{X,Y}(x,y)}{f_{X}(x)}=\begin{cases} \frac{1}{1+x}, & x\in[-1,0],y\geq0
\\\frac{1}{1-x}, & x\in(0,1],y\geq0
\\ 0, & else  \end{cases}$$


### 2.
$$\mathbb{E}(Y|X)= \begin{cases} \int_{0}^{x+1}y\cdot\frac{1}{1+x}dy, & x\in[-1,0]
\\\int_{0}^{1-x}y\cdot\frac{1}{1-x}dy, & x\in(0,1]
\\ 0, & \mbox{else } \end{cases}=\begin{cases}\frac{1+x}{2}, & x\in[-1,0]
\\\frac{1-x}{2}, & x\in(0,1]
\\ 0, & \mbox{else } \end{cases}$$

### 3.
$$\mathbb{E}(\mathbb{E}(Y|X))=\int_{-1}^0\frac{x+1}{2}\cdot (x+1)dx+\int_{0}^1\frac{1-x}{2}\cdot (1-x)dx=
\frac{(x+1)^3}{6}\bigg|_{-1}^0-\frac{(1-x)^3}{6}\bigg|_{0}^1=\frac{1}{3}\\
f_{Y}(y)=\int_{y-1}^{1-y}1dx=2-2\cdot y\\
\mathbb{E}(Y)=\int_{0}^1y\cdot (2-2\cdot y)dy=y^2 \bigg|_{0}^1-\frac{2}{3}\cdot y^3 \bigg|_{0}^1=\frac{1}{3}$$
We get that $\mathbb{E}(\mathbb{E}(Y|X))=\mathbb{E}(Y)$

## Additional

```python
mean1 = [0,1]
cov1 = [[1, -0.7], [-0.7, 2]]
mean2=[7,8]
cov2=[[1,0.3],[0.3,3]]
c= bernoulli.rvs(size=5000,p=0.3)
a=[]
b=[]
for i in c:
    if i==0:
        t,s=np.random.multivariate_normal(mean1, cov1, 1).T
    else:
        t,s=np.random.multivariate_normal(mean2, cov2, 1).T
    a+=list(t)
    b+=list(s)
```

```python
plt.scatter(a,b,c=c,alpha=0.5,s=3)
```

```python
jointplot(a, b,  kind="scatter",s=4,alpha=0.3,marginal_kws=dict(bins=60),color="k")
```

```python
jointplot(a, b,  kind="kde",color="r")
```

```python
ax = plt.axes(projection="3d")
ax.scatter3D(a,b,c,s=0.3 )
```
