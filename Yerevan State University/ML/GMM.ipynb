```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
hw = pd.read_csv('hw.csv')
```

```python
plt.scatter(hw['Height'], hw['Weight'])
```

```python
from sklearn.mixture import GaussianMixture
gmm = GaussianMixture(n_components=4, covariance_type = 'full',random_state=1).fit(hw) 
gmm_labels = gmm.predict(hw)
plt.scatter(hw['Height'], hw['Weight'], c = gmm_labels)
```

```python
data = np.loadtxt('messy.csv', delimiter=',')
```

```python
plt.scatter(data[:,0], data[:,1]);
```

```python
gmm = GaussianMixture(n_components=3, covariance_type = 'full',random_state=1).fit(data) 
gmm_labels = gmm.predict(data)
plt.scatter(data[:,0], data[:,1], c = gmm_labels, alpha = 0.3);
```
